```python
## Deviation Calculation


# 1. Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 2. Import pandas
import pandas as pd

# 3. Load your CSV file
file_path = '/content/drive/MyDrive/SW_IFL/Revised_NationalExposure_Zscore.csv'
df = pd.read_csv(file_path)

# 4. Define years to process
years = ['1980', '1990', '2000', '2010', '2020']

# 5. Initialize a dictionary to store city averages
city_avg_data = {}

# 6. Calculate city averages and deviations
for year in years:
    max_col = f'Tair_Max_{year}'
    min_col = f'Tair_Min_{year}'

    if max_col in df.columns and min_col in df.columns:
        # Calculate city averages (groupby 'ID')
        city_avg = df.groupby('ID')[[max_col, min_col]].mean()
        city_avg.columns = [f'Avg_{max_col}', f'Avg_{min_col}']

        # Store averages for saving later
        city_avg_data[year] = city_avg

        # Compute deviations for census tracts
        df[f'DT_Max_{year}'] = df[max_col] - city_avg[f'Avg_{max_col}'].loc[df['ID']].values
        df[f'DT_Min_{year}'] = df[min_col] - city_avg[f'Avg_{min_col}'].loc[df['ID']].values

# 7. Combine all city averages into one DataFrame
city_avg_combined = pd.concat(city_avg_data.values(), axis=1)

# 8. Save city averages to a new CSV
city_avg_path = '/content/drive/MyDrive/SW_IFL/City_Average_Temperatures.csv'
city_avg_combined.to_csv(city_avg_path)

# 9. Save deviations to the original file (updated)
output_path = '/content/drive/MyDrive/SW_IFL/Revised_NationalExposure_Zscore_WithDeviations.csv'
df.to_csv(output_path, index=False)

# 10. Download both files
from google.colab import files
files.download(city_avg_path)  # City averages
files.download(output_path)    # Deviations

print("✅ Processing complete! Files saved:")
print(f"1. City averages: {city_avg_path}")
print(f"2. Census tract deviations: {output_path}")
```

```python
## Deviation Summary Calculation
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

# Path to your file
file_path = '/content/drive/MyDrive/SW_IFL/NationalExposureULTIMATE.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# List of years we're interested in
years = ['1980', '1990', '2000', '2010', '2020']

# Create a dictionary to specify which aggregations we want for which columns
agg_dict = {}
for year in years:
    agg_dict[f'DT_Max_{year}'] = ['min', 'max']
    agg_dict[f'DT_Min_{year}'] = ['min', 'max']

# Group by ID and aggregate
result_df = df.groupby('ID').agg(agg_dict)

# Flatten the multi-index column names
result_df.columns = ['_'.join(col).strip() for col in result_df.columns.values]
result_df = result_df.reset_index()

# Display the first few rows to verify
print("Aggregated results for first 5 cities:")
print(result_df.head())

# Show summary statistics
print("\nSummary statistics:")
print(f"Number of cities processed: {len(result_df)}")
print(f"Years processed: {years}")
# Save the results to a new CSV file
output_path = '/content/drive/MyDrive/SW_IFL/city_climate_extremes.csv'
result_df.to_csv(output_path, index=False)

print(f"\nResults saved to: {output_path}")
```
